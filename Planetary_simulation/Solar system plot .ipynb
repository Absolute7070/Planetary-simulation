# Plotting the Solar System 

```python
from mpl_toolkits import mplot3d
%matplotlib notebook
import numpy as np
import matplotlib.pyplot as plt 
import pandas as pd 
```

## Dictionaries for saving data 

```python
sun={'x': [], 'y': [], 'z': []}
mercury={'x': [], 'y': [], 'z': []}
venus={'x': [], 'y': [], 'z': []}
earth={'x': [], 'y': [], 'z': []}
mars={'x': [], 'y': [], 'z': []}
jupiter={'x': [], 'y': [], 'z': []}
saturn={'x': [], 'y': [], 'z': []}
uranus={'x': [], 'y': [], 'z': []}
neptune ={'x': [], 'y': [], 'z': []} 

#list of formerly determined objects  
systemlist=[sun, mercury, venus, earth, mars, jupiter, saturn, uranus, neptune] 
```

## Defining function to save data 

```python
def savetodic(components: list):
    global i 
    systemlist[i]['x'].append(components[0])
    systemlist[i]['y'].append(components[1])
    systemlist[i]['z'].append(components[-1])
    if i<len(systemlist)-1:
        i+=1 
    else: 
        i=0
            
```

## Reading file

```python
filename= #output filename 

namelist=['Sun', 'Mercury', 'Venus', 'Earth', 'Mars', 'Jupiter', 'Saturn', 'Uranus', 'Neptune']
```

```python
with open(filename, 'r') as file:
    for row in file:
        #converting rows into list of objects with their 
        # position coordinates 
        row=row.replace('\n', '')
        rowlist=row.split('/')
        rowlist.pop(-1)        #removing the empty element 
        i=0
        #saving data to dictionaries 
        for object in rowlist: 
            components=object.split(' ') 
            components= list(map(lambda x: float(x), components)) #converting strings to float
            savetodic(components)
            
        

            
        
```

## 2D Plot 

```python
#drawing plots 
i=0 
for object in systemlist: 
    plt.plot(object['x'], object['y'], '.', label=namelist[i])
    i+=1 
    
#adjusting axes 
plt.xlabel('x (m)')
plt.ylabel('y (m)') 
plt.legend(loc='best')
plt.show()

```

## 3D plot 

```python
fig=plt.figure()
ax = plt.axes(projection='3d')

#drawing 3d plots 
i=0 
for object in systemlist: 
    ax.scatter3D(object['x'], object['y'], object['z'], '.', label= namelist[i])
    i+=1 
    
#adjusting axes 
ax.set_xlabel('x (m)')
ax.set_ylabel('y (m)')
ax.set_zlabel('z (m)')
ax.legend(loc='best')
plt.show()

```
